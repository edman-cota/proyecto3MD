```python

# Cargar el CSV con todos los datos combinados
df_final = pd.read_csv("datos_combinados.csv")
```

```python
def describe_data(df):
    print(" Descripción del conjunto de datos:")
    print(f"- Número de observaciones: {df.shape[0]}")
    print(f"- Número de variables: {df.shape[1]}")
    print("\n Tipos de variables:")
    print(df.dtypes.value_counts())
    print("\n Valores faltantes:")
    print(df.isnull().sum()[df.isnull().sum() > 0])

describe_data(df_final)
```

```python
import math
import matplotlib.pyplot as plt

def plot_histograms(df):
    num_vars = df.shape[1]
    ncols = 5  # O cualquier número deseado de columnas en la grilla
    nrows = math.ceil(num_vars / ncols)  # Calcula el número de filas necesarias

    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(15, nrows * 2))
    axes = axes.flatten()  # Para iterar sobre los subgráficos fácilmente

    # Iterar sobre las columnas y graficar cada histograma
    for i, col in enumerate(df.columns):
        df[col].hist(ax=axes[i], bins=30, edgecolor='black', alpha=0.7)
        axes[i].set_title(col, fontsize=10)
        axes[i].grid(False)

    # Eliminar subplots sobrantes, en caso de que no se usen todos
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.suptitle(" Histogramas de variables numéricas", fontsize=14, fontweight='bold')
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# Supongamos que df_final es tu DataFrame y seleccionamos las variables numéricas:
plot_histograms(df_final.select_dtypes(include=['number']))

```

```python
for col in df_final.select_dtypes(include=['number']).columns:
    stat, p_value = stats.normaltest(df_final[col].dropna())  # D'Agostino-Pearson
    print(f"\n Test de normalidad para {col}: p-value={p_value:.4f}")
    if p_value < 0.05:
        print("No sigue una distribución normal.")
    else:
        print("Sigue una distribución normal.")

import statsmodels.api as sm

for col in df_final.select_dtypes(include=['number']):  # Solo las 5 primeras para visualizar mejor
    plt.figure(figsize=(6, 4))
    sm.qqplot(df_final[col].dropna(), line="s")
    plt.title(f"Q-Q Plot de {col}")
    plt.show()

```

```python

# Filtrar solo las columnas numéricas
df_numeric = df_final.select_dtypes(include=[float, int])

# Calcular la correlación
corr_matrix = df_numeric.corr()

# Crear la máscara para la parte superior de la matriz
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Ajustar el tamaño de la figura
plt.figure(figsize=(12, 8))

# Crear el mapa de calor
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="RdBu_r",
    center=0,
    mask=mask,
    linewidths=0.5,
    annot_kws={"size": 8}
)

# Ajustar título y etiquetas
plt.title("Mapa de calor de correlación", fontsize=14)
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)

# Mostrar el gráfico
plt.show()

```
